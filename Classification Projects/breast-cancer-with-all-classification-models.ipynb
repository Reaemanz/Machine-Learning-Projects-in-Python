```python
from sklearn import datasets
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import naive_bayes


cancer = datasets.load_breast_cancer()
```

```python
cancer
```

```python
print("Features: ", cancer.feature_names)

print("Labels: ", cancer.target_names)
```

```python
cancer.data.shape
```

```python
print(cancer.data[0:5])
```

```python
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2,random_state=42) # 80% training and 20% test
```

```python
X_train.shape
```

```python
y_train.shape
```

# 1) logistic regression

```python
model0 = LogisticRegression()
model0.fit(X_train,y_train)
y_pred_mod0 = model0.predict(X_test)

print("By using logistic")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_mod0))
print("Precision:",metrics.precision_score(y_test, y_pred_mod0))
print("Recall:",metrics.recall_score(y_test, y_pred_mod0))
```

## 2) KNN

```python
clf =KNeighborsClassifier(n_neighbors=9)
#Train the model using the training sets
clf.fit(X_train, y_train)
#Predict the response for test dataset
y_pred_clf = clf.predict(X_test)

print("By using Knn")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_clf))
print("Precision:",metrics.precision_score(y_test, y_pred_clf))
print("Recall:",metrics.recall_score(y_test, y_pred_clf))
```

## 3.1) Support Vector Machine SVM with polynomial

```python
model1 = svm.SVC(kernel='poly')
model1.fit(X_train, y_train)
y_pred_mod1 = model1.predict(X_test)

print("By using SVM with polynomial")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_mod1))
print("Precision:",metrics.precision_score(y_test, y_pred_mod1))
print("Recall:",metrics.recall_score(y_test, y_pred_mod1))
```

## 3.2) Support Vector Machine SVM with gaussian

```python
model7 = svm.SVC(kernel='rbf')
model7.fit(X_train, y_train)
y_pred_mod7 = model7.predict(X_test)

print("By using SVM with gaussian")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_mod7))
print("Precision:",metrics.precision_score(y_test, y_pred_mod7))
print("Recall:",metrics.recall_score(y_test, y_pred_mod7))
```

## **4) Decision Tree**

```python
model3 = DecisionTreeClassifier()
model3.fit(X_train, y_train)
y_pred_mod3 = model3.predict(X_test)

print("By using Decision Tree")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_mod3))
print("Precision:",metrics.precision_score(y_test, y_pred_mod3))
print("Recall:",metrics.recall_score(y_test, y_pred_mod3))
```

## **5) Random Forest**

```python
model4 = RandomForestClassifier()
model4.fit(X_train, y_train)
y_pred_mod4 = model4.predict(X_test)
print("By using Random Forest")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_mod4))
print("Precision:",metrics.precision_score(y_test, y_pred_mod4))
print("Recall:",metrics.recall_score(y_test, y_pred_mod4))
```

## **6) Naive Bayes**

```python
model5 = naive_bayes.GaussianNB()
model5.fit(X_train, y_train)
y_pred_mod5 = model5.predict(X_test)

print("By using Naive Bayes")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_mod5))
print("Precision:",metrics.precision_score(y_test, y_pred_mod5))
print("Recall:",metrics.recall_score(y_test, y_pred_mod5))
```
